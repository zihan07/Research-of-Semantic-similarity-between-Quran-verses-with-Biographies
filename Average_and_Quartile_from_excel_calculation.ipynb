```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import Workbook
from openpyxl.drawing.image import Image

# Load the Excel file
file_path = '/content/drive/MyDrive/Quran-vs-biography-semantic-similarity-CSE498/U_paraphrase-albert-small-v2/v1 b1.xlsx'  # Replace with your file path
#sheet_name = 'v3 b5'  # Replace with your sheet name
df = pd.read_excel(file_path)

# Assuming the data is in column 'Best Match Score (as percentage)'
# Replace 'Best Match Score (as percentage)' with the actual column name if different
data_column = 'Best Match Score (as percentage)'

# Calculate quartiles and average
Q1 = np.percentile(df[data_column], 25)
Q2 = np.percentile(df[data_column], 50)  # Median
Q3 = np.percentile(df[data_column], 75)
average = np.mean(df[data_column])

# Print the results
print(f"1st Quartile (Q1): {Q1}")
print(f"2nd Quartile (Q2, Median): {Q2}")
print(f"3rd Quartile (Q3): {Q3}")
print(f"Average: {average}")

# Plot the data
plt.figure(figsize=(10, 6))
plt.boxplot(df[data_column], vert=False)
plt.title('Box Plot of Best Match Scores')
plt.xlabel('Best Match Score (as percentage)')
plt.savefig('boxplot.png')  # Save the plot as an image

# Create a new Excel file and save the chart
output_file_path = '/content/drive/MyDrive/Quran-vs-biography-semantic-similarity-CSE498/U_paraphrase-albert-small-v2/C_v1 b1.xlsx'
wb = Workbook()
ws = wb.active

# Add the quartiles and average to the Excel file
ws['A1'] = '1st Quartile (Q1)'
ws['B1'] = Q1
ws['A2'] = '2nd Quartile (Q2, Median)'
ws['B2'] = Q2
ws['A3'] = '3rd Quartile (Q3)'
ws['B3'] = Q3
ws['A4'] = 'Average'
ws['B4'] = average

# Add the chart image to the Excel file
img = Image('boxplot.png')
ws.add_image(img, 'D1')

# Save the Excel file
wb.save(output_file_path)

print(f"Chart and data saved to {output_file_path}")
```
